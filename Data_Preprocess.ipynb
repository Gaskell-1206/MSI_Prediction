# Store Test MSI

```python
# Import dataset
file_path = Path('ex_data/CRC_DX_data_set/CRC_DX_Test_Split/CRC_DX_TEST_MSIMUT - test images')
file_list = list(file_path.glob('**/*'))
len(file_list)
```

```python
file_df = pd.DataFrame()
for file in file_list:
    file_parts = file.parts
    file_dict = {
        'data_set' : file_parts[1].split('_')[0],
        'train_test' : file_parts[2],
        'label' : "MSS" if "MSS" in file_parts[3] else "MSI",
        'img_name': file_parts[4],
        'slice_id' : file_parts[4].split('-',2)[1],
        'subject_id' : file_parts[4].split('-',2)[2].split('.')[0]
    }
    file_df = file_df.append(file_dict,ignore_index=True)
file_df
```

```python
file_df.to_csv('ex_data/CRC_DX_data_set/CRC_DX_Lib/CRC_DX_TEST_MSIMUT.csv',index=False)
```

# CRC_MSIMUT_Test

```python
CRC_MSIMUT_test_df = pd.read_csv('ex_data/CRC_DX_data_set/CRC_DX_CSV/CRC_DX_TEST_MSIMUT.csv')
```

```python
CRC_MSIMUT_test_values = CRC_MSIMUT_test_df['subject_id'].value_counts()
CRC_MSIMUT_test_values
```

```python
len1 = len(CRC_MSIMUT_test_df['subject_id'].value_counts())
len1
```

## Validation

```python
# Validation subjectid
CRC_MSIMUT_val_list = []
for i in range(0,len1,2):
    CRC_MSIMUT_val_list.append(CRC_MSIMUT_test_values.keys()[i])
print(len(CRC_MSIMUT_val_list))
CRC_MSIMUT_val_list
```

```python
# get file_names with strings
import os
src_dir = "ex_data/CRC_DX_data_set/CRC_DX_Test/CRC_DX_TEST_MSIMUT - test images"
def get_file_names_with_strings(str_list, src_dir):
    full_list = os.listdir(src_dir)
    final_list = [nm for ps in str_list for nm in full_list if ps in nm]
    return final_list

CRC_MSIMUT_val_list_path = get_file_names_with_strings(CRC_MSIMUT_val_list, src_dir)
len(CRC_MSIMUT_val_list_path)
```

```python
# Copy files into another list
import shutil
 
# path to source directory
src_dir = src_dir
 
# path to destination directory
dest_dir = 'ex_data/CRC_DX_data_set/CRC_DX_Val/CRC_DX_TEST_MSIMUT - validation images'

for fname in CRC_MSIMUT_val_list_path:
    # copying the files to the
    # destination directory
    shutil.copy2(os.path.join(src_dir,fname), dest_dir)
```

```python
dest_dir = 'ex_data/CRC_DX_data_set/CRC_DX_Val/CRC_DX_TEST_MSIMUT - validation images'
files = os.listdir(dest_dir)
len(files)
```
